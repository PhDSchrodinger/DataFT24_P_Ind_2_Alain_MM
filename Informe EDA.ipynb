```python
# Leeyendo el archivo internet.xlsx
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo Excel y la hoja "Totales VMD"
internet = 'internet.xlsx'
Totales_VMD = 'Totales VMD'
velocidad_df = pd.read_excel(internet, sheet_name=Totales_VMD)

# Mostrar las primeras filas del DataFrame para verificar los datos
print(velocidad_df.head())
```

Como esta presentado por trimestre y años debo agruparlos y sacar la media.

```python
# Eliminar espacios en blanco de los nombres de las columnas
velocidad_df.columns = velocidad_df.columns.str.replace(' ', '')
velocidad_df
```

```python
# Seleccionar las columnas deseadas
velocidad_df = velocidad_df[['Año', 'Trimestre', 'Mbps(Mediadebajada)']]
velocidad_df
```

```python
# Agrupar los datos por año y calcular el promedio de la media de bajada
df_grouped = velocidad_df.groupby('Año')['Mbps(Mediadebajada)'].mean().reset_index()

# Mostrar las primeras filas del DataFrame agrupado para verificar los datos
print(df_grouped.head())
```

Extraer los datos de los años y la media de bajada y calcular el incremento año con año

```python
# Crear el gráfico de barras dobles
fig, ax1 = plt.subplots(figsize=(12, 8))

    ###bars = plt.bar(years, media_bajada, color='skyblue')
# Definir el ancho de las barras y la posición de las barras en el eje x
bar_width = 0.35
index = range(len(years))

# Crear las barras para la media de bajada
bars1 = ax1.bar(index, media_bajada, bar_width, label='Media de Bajada (Mbps)', color='skyblue')

# Crear las barras para los incrementos, desplazadas a la derecha
bars2 = ax1.bar([i + bar_width for i in index], incrementos, bar_width, label='Incremento (Mbps)', color='lightgreen')

# Añadir títulos y etiquetas
ax1.set_title('Media de Bajada e Incremento por Año')
ax1.set_xlabel('Año')
ax1.set_ylabel('Mbps')
ax1.set_xticks([i + bar_width / 2 for i in index])
ax1.set_xticklabels(years)
ax1.legend()

# Añadir los valores de y en cada barra
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom')  # va='bottom' para que el texto esté encima de la barra

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom')  # va='bottom' para que el texto esté encima de la barra

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Filtrar los datos para los años 2023 y 2024
df_filtered = velocidad_df[velocidad_df['Año'].isin([2022, 2023, 2024])]
df_filtered = velocidad_df[
    (velocidad_df['Año'].isin([2023, 2024])) |
    ((velocidad_df['Año'] == 2022) & (velocidad_df['Trimestre'].isin([4])))
]


# Ordenar las filas del año más viejo al más reciente
df_filtered = df_filtered.sort_values(by=['Año', 'Trimestre'])

# Calcular el incremento de velocidad trimestre con trimestre
df_filtered['Incremento'] = df_filtered['Mbps(Mediadebajada)'].diff()

# Calcular el incremento de velocidad trimestre con trimestre en porcentaje
df_filtered['Incremento (%)'] = df_filtered['Mbps(Mediadebajada)'].pct_change() * 100

# Mostrar las primeras filas del DataFrame agrupado para verificar los datos
print(df_filtered)
```

```python

# Filtrar los datos para excluir el año 2022
df_filtered_no_2022 = df_filtered[df_filtered['Año'] != 2022]

# Extraer los datos de los años, trimestres, la media de bajada y los incrementos
years = df_filtered_no_2022['Año']
trimestres = df_filtered_no_2022['Trimestre']
media_bajada = df_filtered_no_2022['Mbps(Mediadebajada)']
incrementos = df_filtered_no_2022['Incremento (%)']

# Crear una nueva columna que combine el año y el trimestre
years_trimestres = years.astype(str) + '_Q' + trimestres.astype(str)

# Crear el gráfico de barras dobles
fig, ax1 = plt.subplots(figsize=(12, 8))

# Definir el ancho de las barras y la posición de las barras en el eje x
bar_width = 0.35
index = range(len(years_trimestres))

# Crear las barras para la media de bajada
bars1 = ax1.bar(index, media_bajada, bar_width, label='MediadeBajada(Mbps)', color='skyblue')

# Crear las barras para los incrementos, desplazadas a la derecha
bars2 = ax1.bar([i + bar_width for i in index], incrementos, bar_width, label='Incremento (%) (Mbps)', color='lightgreen')

# Añadir títulos y etiquetas
ax1.set_title('Media de Bajada e Incremento (%) por Año y Trimestre')
ax1.set_xlabel('Año y Trimestre')
ax1.set_ylabel('Mbps')
ax1.set_xticks([i + bar_width / 2 for i in index])
ax1.set_xticklabels(years_trimestres, rotation=45)
ax1.legend()

# Añadir los valores de y en cada barra
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom')  # va='bottom' para que el texto esté encima de la barra

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 2), va='bottom')  # va='bottom' para que el texto esté encima de la barra

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Leer los datos de la hoja "Totales Accesos Por Tecnología"
file_path = 'internet.xlsx'  
T_A_Tecno_df = pd.read_excel(file_path, sheet_name='Totales Accesos Por Tecnología')

# Agrupar los datos por año y calcular el promedio de los trimestres por tecnología
T_A_Tecno_df_grouped_ADSL = T_A_Tecno_df.groupby('Año')['ADSL'].mean().reset_index()
T_A_Tecno_df_grouped_Cablemodem = T_A_Tecno_df.groupby('Año')['Cablemodem'].mean().reset_index()
T_A_Tecno_df_grouped_F_O = T_A_Tecno_df.groupby('Año')['Fibra óptica'].mean().reset_index()
T_A_Tecno_df_grouped_Wireless = T_A_Tecno_df.groupby('Año')['Wireless'].mean().reset_index()
T_A_Tecno_df_grouped_Otros = T_A_Tecno_df.groupby('Año')['Otros'].mean().reset_index()
T_A_Tecno_df_grouped_Total = T_A_Tecno_df.groupby('Año')['Total'].mean().reset_index()

# Unir todos los DataFrames en uno solo usando merge
T_A_Tecno_df_grouped = T_A_Tecno_df_grouped_ADSL
T_A_Tecno_df_grouped = T_A_Tecno_df_grouped.merge(T_A_Tecno_df_grouped_Cablemodem, on='Año')
T_A_Tecno_df_grouped = T_A_Tecno_df_grouped.merge(T_A_Tecno_df_grouped_F_O, on='Año')
T_A_Tecno_df_grouped = T_A_Tecno_df_grouped.merge(T_A_Tecno_df_grouped_Wireless, on='Año')
T_A_Tecno_df_grouped = T_A_Tecno_df_grouped.merge(T_A_Tecno_df_grouped_Otros, on='Año')
T_A_Tecno_df_grouped = T_A_Tecno_df_grouped.merge(T_A_Tecno_df_grouped_Total, on='Año')

T_A_Tecno_df_grouped
```

```python
# Seleccionar las columnas de interés
columns_of_interest = ['Año', 'ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total']
df_grouped = T_A_Tecno_df_grouped[columns_of_interest]

# Graficar los datos
plt.figure(figsize=(10, 6))
for column in columns_of_interest[1:]:
    plt.plot(df_grouped['Año'], df_grouped[column], marker='o', label=column)

plt.title('Comportamiento de Tecnologías de Acceso a Internet Año con Año')
plt.xlabel('Año')
plt.ylabel('Promedio de Accesos')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Ejemplo de datos
years = [1992, 1993, 1994, ..., 2022]
internet_access = [0.001, 0.002, 0.003, 0.1, 0.1, 0.3, 0.8, 3.3, 7, 9.8, 10.9, 11.9, 16,
                   17.7, 20.9, 25.9, 28.1, 34, 45, 51, 55.8, 59.9, 64.7, 68, 71, 74.3, 77.7, 
                    79.9, 85.5, 87.2, 88.4]  # Porcentajes de acceso

# Crecimiento absoluto año a año
crecimiento_absoluto = [internet_access[i] - internet_access[i-1] for i in range(1, len(internet_access))]

# Tasa de Crecimiento Anual Compuesta (CAGR) entre 1992 y 2022
valor_inicial = internet_access[0]
valor_final = internet_access[-1]
n = len(years) - 1

cagr = (valor_final / valor_inicial) ** (1/n) - 1

print("Crecimiento Absoluto Año a Año:", crecimiento_absoluto)
print("Tasa de Crecimiento Anual Compuesta (CAGR):", cagr)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer las hojas del archivo telefonia_fija.xlsx
telefonia_fija = pd.ExcelFile('telefonia_fija.xlsx')
fijas_accesos_tot = telefonia_fija.parse('Fija_accesos_tot')
fija_ingresos = telefonia_fija.parse('Fija_ingresos')
fija_penetracion_tot = telefonia_fija.parse('Fija_penetracion_tot')

# Leer las hojas del archivo Telefonia_movil.xlsx
telefonia_movil = pd.ExcelFile('Telefonia_movil.xlsx')
movil_penetracion = telefonia_movil.parse('Penetracion')
movil_ingresos = telefonia_movil.parse('Ingresos')

# Graficar los datos de telefonia_fija.xlsx
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.bar(fijas_accesos_tot['Año'], fijas_accesos_tot['Accesos telefonía fija (total)'], label='Accesos Totales')
plt.plot(fija_ingresos['Año'], fija_ingresos['Ingresos (miles de $)'], label='Ingresos')
plt.plot(fija_penetracion_tot['Año'], fija_penetracion_tot['Accesos telefonía fija por cada 100 hab'], label='Penetración Total')
plt.title('Telefonía Fija')
plt.xlabel('Años')
plt.ylabel('Valores')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Agregar cuadrícula detallada
plt.minorticks_on()  # Activar marcas menores
plt.grid(True, which='minor', linestyle=':', linewidth=0.5)  # Agregar cuadrícula menor

# Graficar los datos de Telefonia_movil.xlsx
plt.subplot(2, 1, 2)
plt.plot(movil_penetracion['Año'], movil_penetracion['Accesos por cada 100 hab'], label='Penetración')
plt.plot(movil_ingresos['Año'], movil_ingresos['Ingresos (miles de $)'], label='Ingresos')
plt.title('Telefonía Móvil')
plt.xlabel('Años')
plt.ylabel('Valores')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Agregar cuadrícula detallada
plt.minorticks_on()  # Activar marcas menores
plt.grid(True, which='minor', linestyle=':', linewidth=0.5)  # Agregar cuadrícula menor

# Agregar etiqueta bajo la gráfica
plt.figtext(0.5, 0.01, 'Fuente: Datos de telefonía fija y móvil', ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```
